# Plots PPO and DQN task9.
Simply drop the logs folders into google colab and click "run all"

```python
import pandas as pd
import matplotlib.pyplot as plt
import os
```

```python
# Detects log folders
ppo_folder = "logs_ppo_carracing"
dqn_folder = "logs_dqn_carracing"
```

```python
# Checks if they exist
for folder in [ppo_folder, dqn_folder]:
    if not os.path.exists(folder):
        raise FileNotFoundError(f"{folder} not found.")
```

```python
# load logs
ppo_rewards = pd.read_csv(os.path.join(ppo_folder, "episode_rewards.csv"))
ppo_losses = pd.read_csv(os.path.join(ppo_folder, "losses.csv"))
ppo_policy = pd.read_csv(os.path.join(ppo_folder, "policy_stats.csv"))

dqn_rewards = pd.read_csv(os.path.join(dqn_folder, "episode_rewards.csv"))
dqn_losses = pd.read_csv(os.path.join(dqn_folder, "losses.csv"))
dqn_qstats = pd.read_csv(os.path.join(dqn_folder, "q_stats.csv"))
```

# PPO 

```python
# PPO plot
plt.figure(figsize=(12,6))
plt.plot(ppo_rewards['episode'], ppo_rewards['reward'], color='blue', label="PPO Reward")
plt.xlabel("Episode")
plt.ylabel("Reward")
plt.title("PPO Episode Rewards")
plt.grid(True)
plt.legend()
plt.show()
```

```python
# PPO losses over time
plt.figure(figsize=(12,6))
plt.plot(ppo_losses["update_idx"], ppo_losses["actor_loss"], label="Actor Loss", color="red")
plt.plot(ppo_losses["update_idx"], ppo_losses["critic_loss"], label="Critic Loss", color="green")
plt.plot(ppo_losses["update_idx"], ppo_losses["total_loss"], label="Total Loss", color="purple")
plt.xlabel("Update")
plt.ylabel("Loss")
plt.title("PPO Losses Over Training")
plt.grid(True)
plt.legend()
plt.show()
```

```python
# PPO Entorpy 
plt.figure(figsize=(12,6))
plt.plot(ppo_losses["update_idx"], ppo_losses["entropy"], label="Policy Entropy", color="orange")
plt.xlabel("Update")
plt.ylabel("Entropy")
plt.title("PPO Policy Entropy Over Time")
plt.grid(True)
plt.legend()
plt.show()
```

```python
# Plot mean action output
plt.figure(figsize=(12,6))
plt.plot(ppo_policy["update_idx"], ppo_policy["mean_action"], color="blue")
plt.xlabel("Update")
plt.ylabel("Mean Action Output")
plt.title("PPO Policy Mean Action Over Time")
plt.grid(True)
plt.show()
```

```python
# Plots standard deviation
plt.figure(figsize=(12,6))
plt.plot(ppo_policy["update_idx"], ppo_policy["std_action"], color="red")
plt.xlabel("Update")
plt.ylabel("Std of Action Output")
plt.title("PPO Policy Std Over Time")
plt.grid(True)
plt.show()
```

# DQN

```python
# DQN plot
plt.figure(figsize=(12,6))
plt.plot(dqn_rewards['episode'], dqn_rewards['reward'], color='orange', label="DQN Reward")
plt.xlabel("Episode")
plt.ylabel("Reward")
plt.title("DQN Episode Rewards")
plt.grid(True)
plt.legend()
plt.show()
```

```python
# DQN losses based on update index
plt.figure(figsize=(12,6))
plt.plot(dqn_losses["update_idx"], dqn_losses["loss"], color="red")
plt.xlabel("Training Step")
plt.ylabel("Loss")
plt.title("DQN Loss Over Time")
plt.grid(True)
plt.show()
```

```python
# DQN Q-value metrics over update index. 
plt.figure(figsize=(12,6))
plt.plot(dqn_qstats["update_idx"], dqn_qstats["mean_q"], label="Mean Q", color="blue")
plt.plot(dqn_qstats["update_idx"], dqn_qstats["max_q"], label="Max Q", color="green")
plt.xlabel("Training Step")
plt.ylabel("Q-value")
plt.title("Q-Value Statistics During Training")
plt.legend()
plt.grid(True)
plt.show()
```
